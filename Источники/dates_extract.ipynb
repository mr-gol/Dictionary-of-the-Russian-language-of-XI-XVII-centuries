```python
import os
import re
from bs4 import BeautifulSoup as BS
import pandas as pd
from string import digits
import numpy as np
import lxml
```

```python
alldates_uniq = [] #список, в котором будут уникальные теги <date>, <origdate> из всех 27 томов
```

```python
for volume in range(1, 28):
    f = '/content/oldrus-tei-' + str(volume) + '.xml'
    with open(f, 'r', encoding='utf-8') as xml_doc:
      soup = BS(xml_doc, 'lxml')

    dates = soup.find_all('date')
    origdates = soup.find_all('origdate')

    for i in range(0, len(dates)):
      da = str(dates[i])
      if (da in alldates_uniq):
        pass
      else:
        alldates_uniq.append(da)

    for j in range(0, len(origdates)):
      oda = str(origdates[j])
      if (oda in alldates_uniq):
        pass
      else:
        alldates_uniq.append(oda)
```

```python
with open('unique_dates_origdates_raw.txt', 'a', encoding = 'utf-8') as udod:
    for q in range(0, len(alldates_uniq)):
      udod.write(alldates_uniq[q] + '\n')
```
